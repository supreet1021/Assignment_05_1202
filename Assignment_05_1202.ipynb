```python
def new_visualize(dataset):
    features = np.arange(0,40,1)
    
    dataset_new = pd.read_csv(filepath_or_buffer = dataset)
    
    maxs = dataset_new.apply(np.max)
    plt.plot(features,maxs,label='max')
    
    mins = dataset_new.apply(np.min)
    plt.plot(features,mins,label='min')
    
    avg = dataset_new.apply(np.mean)
    plt.plot(features,avg,label='mean')
    
    plt.legend()
    plt.xlabel('Features')
    plt.ylabel('Values')
    plt.title(dataset)
    plt.show()
```
