```python
def visualize(dataset):
    features = np.arange(0,40,1)
    dataset = pd.read_csv(filepath_or_buffer = dataset)
    plt.figure(figsize=(20, 10))
    plt.subplot(2, 2, 1) 
    plt.plot(features,dataset.min(),label ='Min')
    plt.plot(features,dataset.max(),label ='Max')
    plt.plot(features,dataset.mean(),label ='Average')
    plt.xlabel('feature #')
    plt.legend()
    plt.ylabel("Values")
    plt.show()
```

```python
def identify_issues(dataset):
    dataset = pd.read_csv(filepath_or_buffer = dataset)
    feature_0 = dataset.iloc[:, 0]
    feature_20 = dataset.iloc[:, 20]
    minValuesObj = dataset.min()
    sum_of_mins = sum(minValuesObj)

    
    if(feature_0.max() == 0 and feature_20.max() == 20):
        print("I doubt the integrity of the dataset")
    
    elif( sum_of_mins == 0): 
        print("why is this happening") 
    
    else: 
        print("Dataset looks good") 
    
```
